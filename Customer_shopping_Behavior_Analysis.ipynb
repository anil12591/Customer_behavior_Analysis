```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
labels = ['Young Adults', 'Adult', 'Middle-aged', 'senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly' : 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
from sqlalchemy import create_engine
import urllib.parse

# 1. Setup connection details
username = "postgres"
password = urllib.parse.quote_plus("Anil!@") # Encodes special characters like ! and @
host = "127.0.0.1"  # Using 127.0.0.1 instead of localhost avoids socket errors on Windows
port = "5432"
database = "Customer_behavaior"  # Verify if this matches your pgAdmin spelling!

# 2. Create the engine
engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# 3. Load DataFrame into PostgreSQL
try:
    table_name = "Customer"
    df.to_sql(table_name, engine, if_exists="replace", index=False)
    print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
except Exception as e:
    print(f"Error: {e}")
```
